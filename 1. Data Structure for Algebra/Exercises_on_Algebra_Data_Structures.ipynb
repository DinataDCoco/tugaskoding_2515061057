```python
# selesaikan soal berikut
# what is the transpose of this vector?
#   [
#    [25],
#    [2],
#    [-3],
#    [-23],
#         ]

import numpy as np

transpose_np = np.array([[25], [2], [-3], [-23]])
print(transpose_np)

print()

# transporse dari transpose_np menjadi transpose_t
transpose_t = transpose_np.T
print(f"Hasil Transpose : {transpose_t}")
```
